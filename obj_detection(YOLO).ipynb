## Train the model

```
from ultralytics import YOLO

# Load a pre-trained YOLOv11 model (choose n/s/m/l depending on size)

model = YOLO("yolo11m.pt")
model.train(data="/content/yolo11_project/data.yaml", epochs=50, imgsz=640)
```
